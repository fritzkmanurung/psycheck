### 3. **Penanganan Missing Data**

Beberapa kolom memiliki nilai yang hilang, dan perlu dilakukan penanganan:
- **Profession**, **Academic Pressure**, **Work Pressure**, **CGPA**, **Study Satisfaction**, **Dietary Habits**, **Degree**, dan **Financial Stress** memiliki nilai yang hilang.
- **Strategi Penanganan**: Kita dapat mengisi data yang hilang menggunakan strategi imputasi, seperti menggunakan rata-rata, median, atau nilai paling umum (mode) untuk kolom numerik, dan mengisi kolom kategorikal dengan kategori yang paling sering muncul atau nilai 0.

```python
numerical_vars = ['Age', 'Academic Pressure', 'Work Pressure', 'CGPA', 'Study Satisfaction',
                  'Job Satisfaction', 'Work/Study Hours', 'Financial Stress', 'Depression']

categorical_vars = ['Gender', 'Working Professional or Student', 'Sleep Duration', 'Dietary Habits', 
                    'Have you ever had suicidal thoughts ?', 'Family History of Mental Illness']
```

```python
train_df = train_df[numerical_vars+categorical_vars]
```

# Filter relevant columns

```python
def identify_cat_above30(series):
    counts = series.value_counts()
    return list(counts[counts>=30].index)
```

Alasan kami menggunakan categori yang muncul di atas 30 kali adalah untuk melakuikan analisis pada kategori yang paling sering muncul saja sehingga analisis atau model tidak terganggu oleh kategori langka yang mungkin tidak representatif atau terlalu sedikit untuk memberikan informasi yang bermakna. Kategori yang jarang muncul (misalnya, hanya muncul 1-2 kali) juga bisa jadi tidak memberikan kontribusi signifikan terhadap analisis. Dengan membatasi pada kategori yang lebih sering, kita bisa fokus pada pola yang lebih jelas dan relevan.


```python
levels_to_keep = train_df[categorical_vars].apply(identify_cat_above30, axis=0)
levels_to_keep
```

Dari hasil pemfilteran data:
- sleep duration akan memiliki 4 nilai yaitu ['Less than 5 hours', '5-6 hours', '7-8 hours', 'more than 8 hours']
- dietary habits akan memiliki 3  nilai yaitu ['Unhealthy', 'Moderate', 'Healthy']

```python
for var in categorical_vars:
    train_df = train_df.loc[train_df[var].isin(levels_to_keep[var])]
```

# Model 3: Neural Network (Keras)

```python
model_student = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_students.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])
```

```python
model_student.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history_student = model_student.fit(
    X_train_students, y_train_students,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)
```

```python
test_loss, test_accuracy = model_student.evaluate(X_train_students, y_train_students)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
y_pred_students_nn = model_student.predict(X_test_students)
y_pred_students_nn = (y_pred_students_nn > 0.5).astype(int)
print("\nTest Set Evaluation Student (Neural Network)")
print(classification_report(y_test_students, y_pred_students_nn))
print("Test Accuracy (Neural Network):", accuracy_score(y_test_students, y_pred_students_nn))
```
