````python
numerical_vars = ['Age', 'Academic Pressure', 'Work Pressure', 'CGPA', 'Study Satisfaction',
                  'Job Satisfaction', 'Work/Study Hours', 'Financial Stress']

categorical_vars = ['Gender', 'Working Professional or Student', 'Sleep Duration', 'Dietary Habits', 
                    'Have you ever had suicidal thoughts ?', 'Family History of Mental Illness']
````

````python
def identify_cat_above30(series):
    counts = series.value_counts()
    return list(counts[counts>=30].index)
````

````python
levels_to_keep = df[categorical_vars].apply(identify_cat_above30, axis=0)
levels_to_keep
````

````python
for var in categorical_vars:
    df = df.loc[df[var].isin(levels_to_keep[var])]
````

````python
df.shape
````

````python
fig, ax = plt.subplots(3,2,figsize=(50,50))
for var, subplot in zip(categorical_vars, ax.flatten()):
    df[var].value_counts().plot(kind='bar', ax=subplot, title=var)

    fig.tight_layout
````

````python
correlations = df[numerical_vars].corr()
````

````python
correlations
````

````python
fig, ax = plt.subplots(figsize=(7,5))
sns.heatmap(correlations, ax=ax, cmap='Blues')
````

## Handling Missing Value

````python
total = df.isnull().sum().sort_values(ascending=False)

percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)

missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Missing Percent'])
````

````python
missing_data['Missing Percent'] = missing_data['Missing Percent'].apply(lambda x: x*100)
missing_data.loc[missing_data['Missing Percent'] > 10][:10]
````

### Check Missing Value
- ```Missing Completely at Random``` berarti data hilang secara acak, tidak terkait dengan nilai variabel yang hilang atau variabel lain dalam dataset. Contohnya pada jam tangan pintar, baterai mati tiba-tiba sehingga tidak merekam data detak jantung.

- ```Missing at Random``` adalah kondisi di mana data yang hilang tidak sepenuhnya acak, tetapi probabilitas kehilangan data pada sebuah variabel bergantung pada variabel lain yang diamati. Artinya, ada pola tertentu yang menyebabkan data hilang, tetapi pola tersebut bisa dijelaskan oleh informasi yang ada di variabel lain. Contohnya pada jam tangan pintar, jika data detak jantung tidak direkam karena pengguna sering melepas jam tangan setelah sore hari. Pola kehilangan data (sore hari) dapat diprediksi dari variabel waktu atau kebiasaan pengguna.

- ```Missing Not at Random``` adalah kondisi di mana data yang hilang tidak hanya bergantung pada variabel lain, tetapi secara langsung pada nilai yang hilang itu sendiri. Contohnya pada jam tangan pintar, pengguna mungkin sengaja mematikan sensor detak jantung karena merasa tidak nyaman saat menggunakan fitur ini, sehingga data yang hilang bergantung langsung pada nilai detak jantung itu sendiri atau preferensi pengguna.

Berdasarkan analisis terhadap pola missing value, ditemukan beberapa hubungan yang menjelaskan alasan data hilang pada beberapa variabel. Berikut adalah pengamatan yang diperoleh:

1. **Variabel Academic Pressure, Study Satisfaction, dan CGPA**:
   - Missing value ditemukan untuk semua baris di mana variabel **Working Professional or Student** memiliki nilai **Working Professional**.
   - Hal ini menunjukkan bahwa variabel-variabel tersebut hanya relevan untuk kelompok **Student**, karena **Academic Pressure** dan **Study Satisfaction** umumnya tidak berlaku untuk **Working Professionals**.

2. **Variabel Work Pressure dan Job Satisfaction**:
   - Missing value ditemukan untuk semua baris di mana variabel **Working Professional or Student** memiliki nilai **Student**.
   - Hal ini menunjukkan bahwa variabel-variabel tersebut hanya relevan untuk kelompok **Working Professional**, karena **Work Pressure** dan **Job Satisfaction** umumnya tidak berlaku untuk **Students**.

Berdasarkan pola missing value yang ditemukan, dapat disimpulkan bahwa:

- Missing value dalam dataset ini bersifat **Missing at Random (MAR)**.
- Missing value pada setiap variabel tidak terjadi secara acak, tetapi terkait dengan variabel **Working Professional or Student**.

Tahap penyelesaian masalah yang akan dilakukan adalah:
1. **Pisahkan Data Berdasarkan Kelompok**
2. **Hapus Kolom yang Tidak Relevan untuk Setiap Kelompok**
