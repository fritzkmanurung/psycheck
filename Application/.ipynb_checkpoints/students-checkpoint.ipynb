## Handling Missing Value

````python
total = df.isnull().sum().sort_values(ascending=False)

percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)

missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Missing Percent'])
````

````python
missing_data['Missing Percent'] = missing_data['Missing Percent'].apply(lambda x: x*100)
missing_data.loc[missing_data['Missing Percent'] > 10][:10]
````

### Check Missing Value
- ```Missing Completely at Random``` berarti data hilang secara acak, tidak terkait dengan nilai variabel yang hilang atau variabel lain dalam dataset. Contohnya pada jam tangan pintar, baterai mati tiba-tiba sehingga tidak merekam data detak jantung.

- ```Missing at Random``` adalah kondisi di mana data yang hilang tidak sepenuhnya acak, tetapi probabilitas kehilangan data pada sebuah variabel bergantung pada variabel lain yang diamati. Artinya, ada pola tertentu yang menyebabkan data hilang, tetapi pola tersebut bisa dijelaskan oleh informasi yang ada di variabel lain. Contohnya pada jam tangan pintar, jika data detak jantung tidak direkam karena pengguna sering melepas jam tangan setelah sore hari. Pola kehilangan data (sore hari) dapat diprediksi dari variabel waktu atau kebiasaan pengguna.

- ```Missing Not at Random``` adalah kondisi di mana data yang hilang tidak hanya bergantung pada variabel lain, tetapi secara langsung pada nilai yang hilang itu sendiri. Contohnya pada jam tangan pintar, pengguna mungkin sengaja mematikan sensor detak jantung karena merasa tidak nyaman saat menggunakan fitur ini, sehingga data yang hilang bergantung langsung pada nilai detak jantung itu sendiri atau preferensi pengguna.

Berdasarkan analisis terhadap pola missing value, ditemukan beberapa hubungan yang menjelaskan alasan data hilang pada beberapa variabel. Berikut adalah pengamatan yang diperoleh:

1. **Variabel Academic Pressure, Study Satisfaction, dan CGPA**:
   - Missing value ditemukan untuk semua baris di mana variabel **Working Professional or Student** memiliki nilai **Working Professional**.
   - Hal ini menunjukkan bahwa variabel-variabel tersebut hanya relevan untuk kelompok **Student**, karena **Academic Pressure** dan **Study Satisfaction** umumnya tidak berlaku untuk **Working Professionals**.

2. **Variabel Work Pressure dan Job Satisfaction**:
   - Missing value ditemukan untuk semua baris di mana variabel **Working Professional or Student** memiliki nilai **Student**.
   - Hal ini menunjukkan bahwa variabel-variabel tersebut hanya relevan untuk kelompok **Working Professional**, karena **Work Pressure** dan **Job Satisfaction** umumnya tidak berlaku untuk **Students**.

Berdasarkan pola missing value yang ditemukan, dapat disimpulkan bahwa:

- Missing value dalam dataset ini bersifat **Missing at Random (MAR)**.
- Missing value pada setiap variabel tidak terjadi secara acak, tetapi terkait dengan variabel **Working Professional or Student**.

Tahap penyelesaian masalah yang akan dilakukan adalah:
1. **Pisahkan Data Berdasarkan Kelompok**
2. **Hapus Kolom yang Tidak Relevan untuk Setiap Kelompok**

### Modelling Students

````python
X_students = students_df.drop(['Depression'], axis=1)
y_students = students_df['Depression']
````

````python
X_students
````

````python
y_students
````

````python
from sklearn.preprocessing import LabelEncoder

for col in categorical_vars:
    le = LabelEncoder()
    X_students[col] = le.fit_transform(X_students[col])
````

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
````

````python
X_students = scaler.fit_transform(X_students)
````

````python
X_students
````

````python
# identify imbalance dataset
count_class = y_students.value_counts()
count_class
````

````python
plt.bar(count_class.index, count_class.values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.xticks(count_class.index, ['Class 1', 'Class 0'])
plt.show()
````

````python
from imblearn.over_sampling import SMOTE
````

````python
smote = SMOTE(sampling_strategy='minority')
X_students, y_students = smote.fit_resample(X_students, y_students)
````

````python
# identify imbalance dataset
count_class = y_students.value_counts()
count_class
````

````python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
import numpy as np
````

````python
X_train_students, X_test_students, y_train_students, y_test_students = train_test_split(
    X_students, y_students, test_size=0.2, random_state=42
)
````

````python
print("=== Random Forest Model ===")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
````

````python
cv_scores_rf = cross_val_score(rf_model, X_train_students, y_train_students, cv=5, scoring='accuracy')
print("Cross-Validation Scores (Random Forest):", cv_scores_rf)
print("Mean CV Accuracy (Random Forest):", np.mean(cv_scores_rf))
print("Standard Deviation CV Accuracy (Random Forest):", np.std(cv_scores_rf))
````

````python
rf_model.fit(X_train_students, y_train_students)
````

````python
y_pred_rf = rf_model.predict(X_test_students)
print("\nTest Set Evaluation (Random Forest)")
print(classification_report(y_test_students, y_pred_rf))
print("Test Accuracy (Random Forest):", accuracy_score(y_test_students, y_pred_rf))
````

````python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],  
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
````

````python
print("=== XGBoost Model ===")
xgb_model = XGBClassifier(random_state=42, eval_metric='logloss')
````

````python
grid_search = GridSearchCV(estimator=xgb_model,
                           param_grid=param_grid,
                           cv=3,  # 3-Fold Cross Validation
                           scoring='accuracy',
                           verbose=1,
                           n_jobs=-1)
````

````python
grid_search.fit(X_train_students, y_train_students)
````

````python
print("Best Parameters:", grid_search.best_params_)
````

````python
best_xgb_model = grid_search.best_estimator_
````

````python
y_pred_best_xgb = best_xgb_model.predict(X_test_students)
````

````python
print("\nTest Set Evaluation (Best XGBoost)")
print(classification_report(y_test_students, y_pred_best_xgb))
print("Test Accuracy (Best XGBoost):", accuracy_score(y_test_students, y_pred_best_xgb))
````

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
````

````python
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_students.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])
````

````python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

````python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
````

````python
history = model.fit(
    X_train_students, y_train_students,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)
````

````python
test_loss, test_accuracy = model.evaluate(X_train_students, y_train_students)
print(f"Test Accuracy: {test_accuracy:.4f}")
````
